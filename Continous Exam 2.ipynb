```python
b = 5
c = 3

x_1 = x1
x_2 = np.dot(b, x2)
x_3 = np.dot(c, x3)
epsilon = np.random.normal(size=20)

ans=0
for i, j, k in zip(x_1, x_2, x_3):
    ans = 10 + i + j + k + epsilon
print(ans)
```

```python
B = 1000
N = 20

bootstrap_data = []

for b in range(B):
    
    indices = np.random.choice(N, N, replace=True)
    
    x1_bs, x2_bs, ans_bs = resample(x1, x2, ans, replace=True, n_samples=20)
    
    x_1 = x1_bs[indices]
    x_2 = x2_bs[indices]
    y_b = ans_bs[indices]
    bootstrap_data.append([x_1,x_2,y_b]) 
    
bootstrap_data[999]
```

```python
from sklearn.linear_model import LinearRegression

def boot_fn2(x1,x2,y):
        
    X = np.stack((x1, x2), axis=1)
    m = LinearRegression(fit_intercept=True).fit(X,y)
    
    return m.intercept_, m.coef_
```

```python
coef_list = []
for b in range(B):

    coef_list.append(boot_fn2(bootstrap_data[b][0],bootstrap_data[b][1],bootstrap_data[b][2]))
```

```python
#b0

b_0 = []
b_1 = []
b_2 = []

for b in range(B):

    b_0.append(coef_list[b][0])
    b_1.append(coef_list[b][1][0])
    b_2.append(coef_list[b][1][1])
```

```python
b0_me = np.array(b_0).mean()
b0_std = np.array(b_0).std()

b1_me = np.array(b_1).mean()
b1_std = np.array(b_1).std()

b2_me = np.array(b_2).mean()
b2_std = np.array(b_2).std()

print('b0 estimate & se:', b0_me, b0_std)
print('b1 estimate & se:', b1_me, b1_std)
print('b2 estimate & se:', b2_me, b2_std)
```

from autoregressive model, we would expect the autocorreclation function to exhibits diminishing behavior overtime. Thus, based on the decaying acf, we are likely dealing with a Auto Regressive process. Based on the partial autocorrelation function, we should start with an auto regresssive model with lags 1 and maybe 11.

```python
nile_train = nile[nile['time'].between(1871, 1965)]
nile_test = nile[nile['time'].between(1966, 1970)]
```

```python
ar_1 = AutoReg(nile_train.value, lags = 1).fit()
ar_1.summary()
```

```python
predictions1 = ar_1.predict(start = nile_test.index[0], 
                           end = nile_test.index[-1])

residuals1 = nile_test.value - predictions1

plt.figure(figsize = (10,4))
plt.plot(nile_test.value)
plt.plot(predictions1)

plt.legend(('Data','Predictions'), fontsize = 15)

plt.title('Annual River Flow', fontsize = 20)
plt.ylabel('Value', fontsize = 15)
```

```python
ar_2 = AutoReg(nile_train.value, lags = 11).fit()
ar_2.summary()
```
